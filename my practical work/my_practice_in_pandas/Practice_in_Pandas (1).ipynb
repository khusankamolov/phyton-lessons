```python
import matplotlib.pyplot as plt

daily_revenue = coffee.groupby('Day')['Revenue'].sum()

plt.figure()
daily_revenue.plot(kind='bar')

plt.title('Daily Revenue')
plt.xlabel('Day')
plt.ylabel('Revenue')

plt.show()
```

```python
drink_revenue = coffee.groupby(['Day', 'Coffee Type'])['Revenue'].sum().unstack()

plt.figure()
drink_revenue.plot(kind='bar')

plt.title('Revenue by Coffee Type')
plt.xlabel('Day')
plt.ylabel('Revenue')

plt.show()
```

#1) Quyida berilgan datasetni yuklab olib, ustun nomlarini koʻring va ularni maʼno jihatdan tushunarli shaklga keltiring. `Unnamed: 0` ustunini tashlab yuboring!

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/restaurant_sales.csv"
df = pd.read_csv(url)
df.head()
```

```python
#javobni shu yerda yozing
df.drop(['Unnamed: 0'], axis = 'columns', inplace=True)
df.rename(columns = {'Order ID':'order_id','Customer ID':'customer_id'}, inplace = True)
df.head()
```

```python
#javobni shu yerda yozing
print(df.info())
print(df.describe())
```

#3) Har bir ustundagi boʻsh qiymatlarni aniqlang va ularni toʻldirish strategiyasini ishlab chiqing.

```python
#javobni shu yerda yozing
df.isnull().sum()
```

```python
df.Category.value_counts()
#df.dropna(subset = ['Category'], inplace = True)
```

```python
df[df.Category.isnull()]
```

```python
print(int(df.Quantity.mean()))
int(df['Order Total'].mean())
```

```python
df.fillna({'Quantity' : int(df.Quantity.mean())}, inplace = True)
df['Order Total'] = np.abs(df.Price*df.Quantity)
df.sample(10)
```

```python
df['Payment Method'].value_counts()
df['Payment Method'] = df['Payment Method'].ffill()
```

```python
df.isnull().sum()
```

```python
df.groupby(['Category','Price'])['Item'].count()
```

```python
df[(df.Price < 0) & (df.Category == "Drinks")] = df[(df.Price < 0) & (df.Category == "Drinks")].fillna({'Item':df[(df.Price == 1) & (df.Category == "Drinks")].iat[0,3]})
```

```python
int(df[df.Category =='Drinks'].Price.mean())
```

```python
df.columns[3]
```

```python
df[(df.Price < 0) & (df.Category == "Drinks")].head()
```

```python
lis = df.Category.unique()
x = lis.tolist()
x.remove('Drinks')
x
```

```python
df[df.Price<0].Price.unique()
```

```python
for idx in x[1:]:
  df[(df.Price < 0) & (df.Category == idx)] = df[(df.Price < 0) & (df.Category == idx)].fillna({'Item' : df[(df.Price == int(df[df.Category == idx].Price.mean())) & (df.Category == idx)].iat[1, 3]})
  df.loc[
    (df.Price < 0) & (df.Category == idx),
    'Price'
] = int(df[df.Category == idx].Price.mean())
```

```python
df[(df.Price < 0) & (df.Category == 'Side Dishes')].isnull().sum()


```

```python
df.isnull().sum()
```

```python
df[df.Category.notnull()]
```

```python
df.loc[df.Category.notnull(), 'Item'] = df[df.Category.notnull()].groupby(['Category', 'Price'])['Item'].transform(
        lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x

)
```

```python
df[df.Category.isnull()]
```
